Càlculo de una tasa de crecimiento anual (lineal y compuesta) con y sin fin función

```
city_name="Instanbul,Turkey"

pop_1927=691000
pop_2017=15029231
pop_change= pop_2017-pop_1927

percentage_gr = (pop_change/pop_1927) * 100

annual_gr=percentage_gr/(2017-1927)

def population_growth(year_one,year_two,population_one,population_two):
  
  pop_change= population_two-population_one
  percentage_gr = (pop_change/population_one) * 100
  growth_rate=percentage_gr/(year_two-year_one)
  return growth_rate

print(annual_gr)

set_one=population_growth(1927,2017,691000,15029231)

print(set_one)

def population_growth2(year_one,year_two,population_one,population_two):

  growth_rate2=pow((population_two/population_one),1/(year_two-year_one))-1

  return growth_rate2

set_two=population_growth2(1927,2017,691000,15029231)

print(set_two)

report="The population grew from " + str(pop_1927) +" to "+ str(pop_2017)+ "with a total change of " + str(pop_change)

print(report)

  
  

```

List comprehension

```
#[operation for item in list]

temperature=[5,12,23,4,-4]
temperature_adjustment=[temp+10 for temp in temperature]
print(temperature_adjustment)

# double
nums = [4, 8, 15, 16, 23, 42]
double_nums=[num*2 for num in nums]
print(double_nums)

#Range and square
nums = range(11)
squares=[num**2 for num in nums]
print(nums)
print(squares)

# even or odd
# % es el modulo de la división. Al aplicar %2 se verifica si el núnero es par o impar. 
nums = [4, 8, 15, 16, 23, 42]
parity=[num%2 for num in nums]
print(parity)

# adding strings
names = ["Elaine", "George", "Jerry", "Cosmo"]
greetings=["Hello, "+name for name in names]
print(greetings)

# First character
names = ["Elaine", "George", "Jerry", "Cosmo"]
first_character=[name[0] for name in names]
print(first_character)

#Product
nested_lists = [[4, 8], [15, 16], [23, 42]]
product=[num1*num2 for (num1,num2) in nested_lists]
print(product)

# zip para unir dos listas
a = [1.0, 2.0, 3.0]
b = [4.0, 5.0, 6.0]

sums=[num1+num2 for (num1,num2) in zip(a,b)]
print(zip(a,b))
print(list(zip(a,b)))
print(sums)

# str y zip
names = ["Shilah", "Arya", "Kele"]
ages = [14, 9, 35]
users = ["Name: " + name + ", Age: " + str(age) for (name, age) in zip(names, ages) ]
print (users)
```

```

from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]


# create your figure here
plt.figure(figsize=(12, 8))

ax1=plt.subplot(1,2,1)
x_values=range(len(months))
plt.plot(x_values,visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visitors')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title('Pages visits per month')

ax2=plt.subplot(1,2,2)
plt.plot(x_values,key_limes_per_month,color='green', marker='o')
plt.plot(x_values,persian_limes_per_month,color='orange', marker='s')
plt.plot(x_values,blood_limes_per_month, color='purple',marker='o')
plt.legend(['Key limes','Persian limes','Blood limes'])
plt.xlabel('Month')
plt.ylabel('Limes sold')
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.title('Limes sales per month')

plt.show()

plt.savefig('Limes_graph.png')
```

```
from matplotlib import pyplot as plt

months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]

#your work here
plt.plot(months,revenue)
ax=plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)

y_lower=[0.9*i for i in revenue]
y_upper=[1.1*i for i in revenue]
plt.fill_between(months,y_lower,y_upper,alpha=0.2)
plt.show()
```
