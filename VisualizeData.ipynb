```
city_name="Instanbul,Turkey"

pop_1927=691000
pop_2017=15029231
pop_change= pop_2017-pop_1927

percentage_gr = (pop_change/pop_1927) * 100

annual_gr=percentage_gr/(2017-1927)

def population_growth(year_one,year_two,population_one,population_two):
  
  pop_change= population_two-population_one
  percentage_gr = (pop_change/population_one) * 100
  growth_rate=percentage_gr/(year_two-year_one)
  return growth_rate

print(annual_gr)

set_one=population_growth(1927,2017,691000,15029231)

print(set_one)

def population_growth2(year_one,year_two,population_one,population_two):

  growth_rate2=pow((population_two/population_one),1/(year_two-year_one))-1

  return growth_rate2

set_two=population_growth2(1927,2017,691000,15029231)

print(set_two)

report="The population grew from " + str(pop_1927) +" to "+ str(pop_2017)+ "with a total change of " + str(pop_change)

print(report)

  
  

```

```
#[operation for item in list]

temperature=[5,12,23,4,-4]
temperature_adjustment=[temp+10 for temp in temperature]
print(temperature_adjustment)

# double
nums = [4, 8, 15, 16, 23, 42]
double_nums=[num*2 for num in nums]
print(double_nums)

#Range and square
nums = range(11)
squares=[num**2 for num in nums]
print(nums)
print(squares)

# even or odd
# % es el modulo de la división. Al aplicar %2 se verifica si el núnero es par o impar. 
nums = [4, 8, 15, 16, 23, 42]
parity=[num%2 for num in nums]
print(parity)

# adding strings
names = ["Elaine", "George", "Jerry", "Cosmo"]
greetings=["Hello, "+name for name in names]
print(greetings)

# First character
names = ["Elaine", "George", "Jerry", "Cosmo"]
first_character=[name[0] for name in names]
print(first_character)

#Product
nested_lists = [[4, 8], [15, 16], [23, 42]]
product=[num1*num2 for (num1,num2) in nested_lists]
print(product)

# zip para unir dos listas
a = [1.0, 2.0, 3.0]
b = [4.0, 5.0, 6.0]

sums=[num1+num2 for (num1,num2) in zip(a,b)]
print(zip(a,b))
print(list(zip(a,b)))
print(sums)

# str y zip
names = ["Shilah", "Arya", "Kele"]
ages = [14, 9, 35]
users = ["Name: " + name + ", Age: " + str(age) for (name, age) in zip(names, ages) ]
print (users)
```

```
# Seleccionar varias filas
#tabla.iloc[3:7] would select all rows starting at the 3rd row and up to but not including the 7th row (i.e., the 3rd row, 4th row, 5th row, and 6th row)
#tabla.iloc[:4] would select all rows up to, but not including the 4th row (i.e., the 0th, 1st, 2nd, and 3rd rows)
#tabla.iloc[-3:] would select the rows starting at the 3rd to last row and up to and including the final row

import pandas as pd

df = pd.DataFrame([
  ['January', 100, 100, 23, 100],
  ['February', 51, 45, 145, 45],
  ['March', 81, 96, 65, 96],
  ['April', 80, 80, 54, 180],
  ['May', 51, 54, 54, 154],
  ['June', 112, 109, 79, 129]],
  columns=['month', 'clinic_east',
           'clinic_north', 'clinic_south',
           'clinic_west']
)

print(df)

april_may_june=df.iloc[-3:]
print(april_may_june)

print(df.iloc[:4])
print(df.iloc[1:5])
```

```
import pandas as pd

def lowercase(my_string): # se define una nueva función en python para poner todas las palabras en minuscula
  return my_string.lower()

df = pd.DataFrame([
  ['JOHN SMITH', 'john.smith@gmail.com'],
  ['Jane Doe', 'jdoe@yahoo.com'],
  ['joe schmo', 'joeschmo@hotmail.com']
],
  columns=['Name', 'Email']
)

df['Lowercase Name']=df.Name.apply(lowercase) # con la función apply se aplica a la columna Name de la tabla df la función lowercase

print(df)
```

```
import pandas as pd

def lowercase(my_string): # se define una nueva función en python para poner todas las palabras en minuscula
  return my_string.lower()

df = pd.DataFrame([
  ['JOHN SMITH', 'john.smith@gmail.com'],
  ['Jane Doe', 'jdoe@yahoo.com'],
  ['joe schmo', 'joeschmo@hotmail.com']
],
  columns=['Name', 'Email']
)

df['Lowercase Name']=df.Name.apply(lowercase) # con la función apply se aplica a la columna Name de la tabla df la función lowercase

print(df)

df.columns=['Name','Email','Lower'] # con la función .columns se cambia el nombre de las columnas. Tener mucho cuidado del orden. 

print(df)
```

```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns



# Take in the data from the CSVs as NumPy arrays:
from google.colab import files
files.upload()
set_one = np.genfromtxt("dataset1.csv", delimiter=",")

from google.colab import files
files.upload()
set_two = np.genfromtxt("dataset2.csv", delimiter=",")

from google.colab import files
files.upload()
set_three = np.genfromtxt("dataset3.csv", delimiter=",")

from google.colab import files
files.upload()
set_four = np.genfromtxt("dataset4.csv", delimiter=",")

# Creating a Pandas DataFrame:
n=500
df = pd.DataFrame({
    "label": ["set_one"] * n + ["set_two"] * n + ["set_three"] * n + ["set_four"] * n,
    "value": np.concatenate([set_one, set_two, set_three, set_four])
})

# Setting styles:
sns.set_style("darkgrid")
sns.set_palette("pastel")

print(df)

# Add your code below:

sns.barplot(data=df, x="label", y="value")
plt.show()

sns.kdeplot(set_one, shade=True)
sns.kdeplot(set_two, shade=True)
sns.kdeplot(set_three, shade=True)
sns.kdeplot(set_four, shade=True)

plt.show()

sns.boxplot(data=df, x="label", y="value")
plt.show()

sns.violinplot(data=df,x="label",y="value")
plt.show()




```
